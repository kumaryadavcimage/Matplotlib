#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.


#### solve

Matplotlib is a comprehensive 2D plotting library for Python. It is widely used for creating static, animated, and interactive visualizations in Python. Matplotlib is often used in conjunction with NumPy for numerical operations and Pandas for handling data.

Why Matplotlib is used:

Versatility: Matplotlib provides a wide range of plots and charts, making it suitable for various data visualization tasks.

Customization: Users have fine-grained control over the appearance of plots, allowing them to customize colors, styles, labels, and more.

Publication-quality Graphics: Matplotlib produces high-quality visualizations suitable for publication and presentation.

Integration: It integrates well with other scientific libraries in the Python ecosystem, such as NumPy and Pandas.

Extensibility: Users can build on top of Matplotlib to create complex visualizations and dashboards.

a.Five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot (plt.plot): Used to visualize the relationship between two variables by connecting data points with lines.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### 
b.Scatter Plot (plt.scatter): Used to display individual data points on a two-dimensional space.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

####
c.Bar Plot (plt.bar): Used to compare categorical data by displaying rectangular bars of different heights.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 20]

plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

####
d.Histogram (plt.hist): Used to represent the distribution of a continuous variable through bins.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

####
e.Pie Chart (plt.pie): Used to display the contribution of each category to the whole.

```python
import matplotlib.pyplot as plt

sizes = [30, 20, 15, 35]
labels = ['Category A', 'Category B', 'Category C', 'Category D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

 #### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

#### solve

A scatter plot is a type of data visualization that displays individual data points on a two-dimensional plane. Each point represents the values of two variables, and the position of the point is determined by the values of these variables. Scatter plots are useful for visualizing relationships between two continuous variables and identifying patterns or trends in the data.

Here's how you can generate a scatter plot using the given code:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, color='blue', alpha=0.7)  # 'alpha' controls the transparency of points

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

####
In this example:

plt.scatter(x, y, color='blue', alpha=0.7) creates a scatter plot with blue points and adjusts the transparency using alpha.

plt.title('Scatter Plot of Generated Data') adds a title to the plot.

plt.xlabel('X-axis') and plt.ylabel('Y-axis') add labels to the x-axis and y-axis, respectively.

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#### solve
The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid, specifying the number of rows and columns. This is useful when you want to compare different datasets or visualize multiple aspects of your data in a single figure.

Here's an example of how to use the subplot() function to draw four line plots using the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(12, 8))

# Line plot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1', color='blue')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Line plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2', color='green')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Line plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3', color='red')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Line plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4', color='purple')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

####
In this example:

plt.subplot(2, 2, 1) specifies the first subplot in a 2x2 grid.

plt.tight_layout() adjusts the layout for better spacing.

plt.legend() adds a legend to each plot

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

#### solve

A bar plot is a data visualization technique that represents categorical data with rectangular bars. The length of each bar is proportional to the values it represents. Bar plots are useful for comparing the values of different categories or groups.

Why bar plots are used:

Comparing values: Bar plots make it easy to visually compare the magnitudes of different categories.

Categorical data: They are effective for visualizing data that can be categorized into distinct groups.

Showing trends: Bar plots can show trends or patterns in the data.

Now, let's create a bar plot and a horizontal bar plot using the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Bar Plot of Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightcoral')
plt.title('Horizontal Bar Plot of Company Profits')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.show()
```

####
In the above code:

plt.bar(company, profit, color='skyblue') creates a vertical bar plot.

plt.barh(company, profit, color='lightcoral') creates a horizontal bar plot.

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

#### solve

A box plot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, such as the minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. The box represents the interquartile range (IQR), and the "whiskers" extend to show the range of the data, excluding outliers.

Why box plots are used:

Visualizing distribution: Box plots provide a concise summary of the distribution of a dataset, making it easy to identify key statistics.

Identifying outliers: The whiskers of a box plot help identify potential outliers in the data.

Comparing distributions: Box plots are useful for comparing the distributions of different datasets.

Now, let's create a box plot using the provided data:

####
In this example:

plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], notch=True, vert=True, patch_artist=True) creates a box plot for two datasets with specified labels.

notch=True adds a notch to the median line.

vert=True creates a vertical box plot.

patch_artist=True fills the box with color for better visualization.
